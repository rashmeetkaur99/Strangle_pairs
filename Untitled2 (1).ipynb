```python
# Extracting only BankNifty data
BN =df.query('Ticker.str.contains("BANKNIFTY")', engine='python')
BN
```

```python
Underlyings = BN.query('Ticker.str.contains("FUT")', engine='python')
Underlyings = Underlyings.drop(['Open','High','Low','Volume','Open Interest'], axis=1)
Underlyings['Date']=[f.rsplit("Y", 1)[1].split("F", 1)[0] for f in Underlyings.Ticker]
Underlyings.timestamp =  pd.to_datetime(Underlyings['timestamp'])
Underlyings.Date = pd.to_datetime(Underlyings['Date'])
Underlyings
```

```python
Underlyings = Underlyings.sort_values('Date').drop_duplicates('timestamp')
Underlyings
```

```python
BN['Underlying'] = [f.rsplit("22", 1)[1].split(".", 1)[0] for f in BN.Ticker]
BN = BN.drop(['Open','High','Low','Volume','Open Interest'], axis =1)
# Extracting strike price and option type from the ticker of Bank Nifty data
BN['Strike_Price'] = BN['Underlying'].str.replace('([A-Z]+)', '')
BN['Option_Type'] = BN['Underlying'].str.extract('([A-Z]+)')
BN.head()
```

```python
BN = BN.sort_values(by = 'timestamp', ascending = True)
```

```python
CallOption =BN.query('Option_Type.str.contains("CE")', engine='python')
CallOption.head()
```

```python
PutOption = BN.query('Option_Type.str.contains("PE")', engine='python')
PutOption.head()
```

```python
New_data = pd.merge(CallOption, PutOption, on='timestamp', how='inner')
New_data = New_data[['Ticker_x', 'Close_x', 'Underlying_x', 'Strike_Price_x','Option_Type_x', 'timestamp','Ticker_y','Close_y','Underlying_y','Strike_Price_y','Option_Type_y']]
New_data
```

```python
New_data.timestamp = pd.to_datetime(New_data['timestamp'])
```

```python
New_data['Strangle_pairs'] = New_data['Underlying_y'] + '-' + New_data['Underlying_x'] 
New_data.info()
```

```python
New_data.Strike_Pirce_x = New_data['Strike_Price_x'] = New_data['Strike_Price_x'].astype(np.int64)
New_data.Strike_Price_y = New_data['Strike_Price_y'] = New_data['Strike_Price_y'].astype(np.int64)
```

```python
diff =abs(New_data.Strike_Price_y - New_data.Strike_Price_x).apply(int)
New_data['diff'] = diff
New_data.head()
```

```python
New_data.drop(New_data[New_data['diff'] !=1000].index, inplace = True)
New_data['Strangle_Price'] = New_data.Close_x+New_data.Close_y
New_data
```

```python
Min_price = pd.DataFrame(New_data.groupby('timestamp').min()['Strangle_Price'])
Min_price
```

```python
New_data2 = pd.merge(New_data, Min_price, on=['timestamp','Strangle_Price'], how ='inner')
Final_data = New_data2.drop(['Ticker_x','Underlying_x','Strike_Price_x','Option_Type_x','Ticker_y','Underlying_y','Strike_Price_y','Option_Type_y'], axis =1)
Final_data['time'] = Final_data['timestamp'].dt.time
Final_data = Final_data[['timestamp','time','Close_x','Close_y','Strangle_pairs','diff','Strangle_Price']]
Final_data
```

```python
Final_data2 = pd.merge(Underlyings, Final_data, how = 'outer')
Final_data2 = Final_data2.sort_values('time').reset_index()
Final_data2
```

```python
Final_data2.plot(y='Strangle_Price', x ='time')	
```

```python
Adjusted_price = 0.01*Final_data2.Strangle_Price.shift(1)
```

```python
Final_data2['Adj_factor'] = np.where(Final_data2.Strangle_pairs != Final_data2.Strangle_pairs.shift(1), Adjusted_price, 0)
Final_data2['Decay'] = (Final_data2.Strangle_Price-Final_data2.Strangle_Price.shift(1))
```

```python
import math
Final_data2['Final_strangle'] = math.nan
Final_data2['Final_strangle'][0] = Final_data2.Strangle_Price[0]
```

```python
for i in range(len(Final_data2.Final_strangle)):
    if Final_data2.Final_strangle[i] == 0:
       Final_data2.Final_strangle[i+1] = Final_data2.Final_strangle[i] + Final_data2.Adj_factor[i+1] + Final_data2.Decay[i+1]
Final_data2
```

1. Close: The price of the underlying
2. Close_x: Premimum of call
3. Close_y: Premium of y
4. Strangle_Price: Close_x+Close_y
